```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

#load the csv
df = pd.read_csv('penglingsExt.csv')

#define colors for the species.
species_colors = {
    'Adelie': 'orange',
    'Chinstrap': 'purple',
    'Gentoo': 'teal'
}

#average them out bc theres too many :D
avgData = df.groupby('species').agg ({
    'flipper_length_mm': 'mean',
    'body_mass_g': 'mean',
    'bill_length_mm': 'mean'
})

#create a scatter plot for each species
fig, ax = plt.subplots(figsize=(9, 4.5))
for species, group in df.groupby('species'):
    ax.scatter(group['flipper_length_mm'], 
               group['body_mass_g'], 
               s= group['bill_length_mm'] / 5, #size of the marker there are so many data points is there even a point.
               c= species_colors[species], 
               label= species, 
               alpha=0.6, 
               edgecolors='w')

#set the title and labels
ax.set_title('Average Penguin Size by Species')
ax.set_xlabel('Flipper Length (mm)')
ax.set_ylabel('Body Mass (g)')

#create a legend!
legend_elements = [Line2D([0], [0], marker='o', color='w', label=species,
                          markerfacecolor=color, markersize=10)
                   for species, color in species_colors.items()]
ax.legend(handles=legend_elements, title='Species')

#show the plot :)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#load the csv
df = pd.read_csv('penglingsExt.csv')

#define colors for the species.
species_colors = {
    'Adelie': 'orange',
    'Chinstrap': 'purple',
    'Gentoo': 'teal'
}

#calculate average flipper length for each species
averageBillDepth = df.groupby('species')['bill_depth_mm'].mean()

#create a bar graph
fig, ax = plt.subplots(figsize=(4, 5))
bars = ax.bar(averageBillDepth.index, averageBillDepth.values, 
              color=[species_colors[species] for species in averageBillDepth.index], width = 0.4)

#set the title and labels
ax.set_title('Bill Depth by Penguin Species (Average)')
ax.set_xlabel('Species')
ax.set_ylabel('Average Bill Depth (mm)')

#add the value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black')

#show the plot
plt.show()
```

```python
import pandas as pd
import plotly.express as px

#load the CSV
df = pd.read_csv('penglingsExt.csv')

#group and count occurrences
dGrp = df.groupby(['diet', 'health_metrics', 'species']).size().reset_index(name='counts')

#create the bubble chart using Plotly Express
fig = px.scatter(dGrp, 
                 x='diet', 
                 y='species', 
                 size='counts',  #use counts as the size of bubbles
                 color='health_metrics',  #differentiate by species
                 hover_name='health_metrics',  #tooltips
                 title='Health Metrics and Diet by Species',
                 labels={'diet': 'Diet', 'species': 'Species', 'health_metrics': 'Health Metrics'},  #axes
                 size_max=60,  #max bubble size
                 width = 900,
                 height = 800) 

#show the plot
fig.show()
```
